```python
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(book_pivot_table)  # Training using cosine similarity
```

```python
def recommend_cosine(book_name):
    index=np.where(book_pivot_table.index==book_name)[0][0]
    print(index)
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    print(similar_items)
    for i in similar_items:
        book_title = book_pivot_table.index[i[0]]
        print(book_title)
```

```python
tsvd = TruncatedSVD(n_components=200, random_state=42) # reduce the dimentionality - similar to PCA
book_pivot_table_tsvd = tsvd.fit_transform(book_pivot_table)
```

```python
def recommend_k(book_name):
    index=np.where(book_pivot_table.index==book_name)[0][0]
    print(index)
    dist, sugg = model.kneighbors(book_pivot_table.iloc[index, :].values.reshape(1,-1),n_neighbors=6)
    print(dist)
    print(sugg)
    for i in range(len(sugg[0])):
        print(book_pivot_table.index[sugg[0][i]])
```

```python
df_grouped = new_ratings.groupby(['ISBN', 'Book-Title']).agg(
    total_ratings=('Book-Rating', 'size'),  # Number of ratings (weight)
    sum_weighted_ratings=('Book-Rating', 'sum')  # Sum of all ratings
).reset_index()

# Calculate the weighted average (sum of ratings / total ratings)
df_grouped['weighted_avg_rating'] = df_grouped['sum_weighted_ratings'] / df_grouped['total_ratings']

# Sort by the weighted average rating in descending order
df_sorted = df_grouped.sort_values(by='weighted_avg_rating', ascending=False)

# Display the result
print(df_sorted[['Book-Title', 'weighted_avg_rating']])
```
